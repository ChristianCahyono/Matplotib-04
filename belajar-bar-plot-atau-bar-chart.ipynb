```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Simple Bar Plot
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#Pengaturan grid dan color

data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data, 
        color='red', #Berungsi untuk mengatur warna
        alpha=0.25) #Berfungsi untuk mengatur tingkat transparansi bar plot
plt.grid(linestyle='--', #Mengatur tipe garis
         linewidth=2, #Untuk mengatur tingkat ketebalan garis
         axis='y', #pilihan: x,y,both. Karena disini axis ='y', maka garisnya akan berasosiasi dengan sumbu y
         alpha=0.50) #Untuk mengatur tingkat transparansi garis(semakin tinggi semakin cerah)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
#Horizontal Bar Plot 

data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.barh(kategori,data)
#plt.barh(kategori[::1], data[::1]) berfungsi untuk membalik urutan list

plt.grid(linestyle='--', #mengatur tipe garis
         linewidth= 1, #mengatur tingkat ketebalan garis
         axis='x', #pilihan: x,y,both. Karena disini axis ='x', maka garisnya akan berasosiasi dengan sumbu x
         alpha=0.75) #mengatur tingkat transparansi garis

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Horisontal Bar Plot')

plt.show()
```

```python
#Stacked Bar Plot

data1 = [25,85,75,40,60]
data2 = [45,35,20,55,10]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label= 'Data 1')
#plt.bar(kategori,data2,label= 'Data 2') 
plt.bar(kategori,data2,label= 'Data 2',bottom=data1) #parameter bottom menspesifikasikan dataset ke-2 ditempatkan diatas dataset ke-1. 

plt.legend() #mengikuti nama label

plt.grid(linestyle='--', 
         linewidth=2, 
         axis='y', 
         alpha=0.75) 

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```
